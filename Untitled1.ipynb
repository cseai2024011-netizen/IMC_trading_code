```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("prices_round_0_day_-1.csv",sep=";")
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
df.head()
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"],unit="ms")
df["mid_price"] = pd.to_numeric(df["mid_price"],errors="coerce")
df = df.dropna(subset=["timestamp","mid_price"])
```

```python
df=df.sort_values("timestamp")
```

```python
df.head(5)
```

```python
plt.plot(df["timestamp"],df["mid_price"])
plt.xlabel("Time")
plt.ylabel("Mid price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Take every 50th row (adjust as needed)
data_sampled = df.iloc[::50]

plt.plot(data_sampled["timestamp"], data_sampled["mid_price"])

plt.xticks(rotation=45)
plt.show()
```

```python
# Select a small slice
subset = df.iloc[:500]

plt.plot(subset["timestamp"], subset["mid_price"])

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter ONE product
data_one = df[df["product"] == "TOMATOES"]

# Sort
data_one = data_one.sort_values("timestamp")

# Plot
plt.plot(data_one["timestamp"], data_one["mid_price"])

plt.xlabel("Time")
plt.ylabel("Mid Price")
plt.title("AMETHYSTS Price")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df["ma"] = df["mid_price"].rolling(window=20).mean()
df["signal"] = "HOLD"

df.loc[df["mid_price"] > df["ma"],"signal"] = "BUY"
df.loc[df["mid_price"] < df["ma"],"signal"] = "SELL"
```

```python
buy = df[df["signal"]=="BUY"]
sell = df[df["signal"]=="SELL"]
plt.plot(df["timestamp"],df["mid_price"],label = "Price")
plt.plot(df["timestamp"],df["ma"],label="MA")

plt.scatter(buy["timestamp"],buy["mid_price"],marker="^")
plt.scatter(sell["timestamp"],sell["mid_price"],marker="v")

plt.legend()
plt.xticks(rotation=45)
plt.show()
```
